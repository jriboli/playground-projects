```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod1.csv"
df = pd.read_csv(filepath)
```

```python
df.head()
```

```python
# Standard Line Plot
plt.plot(df['CPU_core'],df['Price'])
```

```python
# Scatter Plot 
plt.scatter(df['Weight_kg'],df['Price'])
```

```python
# Histogram Plot
price_bins = np.linspace(min(df['Price']), max(df['Price']), 6)
print("Bins: ", price_bins)
bin_labels = ["Very Low", "Low", "Average", "High", "Very High"]
df['Price_Binned'] = pd.cut(df['Price'], price_bins, labels=bin_labels, include_lowest=True)
print("Counts: ",  df['Price_Binned'].value_counts())

plt.hist(df['Price'], price_bins)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Histogram of Pricing")
```
