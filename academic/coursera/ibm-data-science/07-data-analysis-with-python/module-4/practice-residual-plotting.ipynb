### Example of a Residual Plot in Python

Let's train a linear regression model, calculate residuals, and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate sample data
np.random.seed(42)
X = np.linspace(0, 10, 50).reshape(-1, 1)
y = 3 * X.squeeze() + 7 + np.random.normal(0, 2, X.shape[0])  # True function with noise

# Train a linear model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Compute residuals
residuals = y - y_pred

# Plot residuals
plt.scatter(X, residuals, color="blue", alpha=0.6)
plt.axhline(y=0, color='r', linestyle='dashed')  # Reference line at y=0
plt.xlabel("X values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

### How to Interpret the Residual Plot
##### 1. Random scatter around zero (Good Fit)
- No clear pattern means a linear model is appropriate
##### 2. Curved pattern (non-linearity)
- Suggests a non-linear relationship. Consider `Polynomial Regression` instead.
##### 3. Funnel Shape (heteroscedasticity)
- Variance of residuals increases or decreases with X.
- The model struggles with certain ranges of X.
- Consider transforming the data or using a different model.

### Example: Residuals indicating a poor linear fit

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate non-linear (quadratic) data
np.random.seed(42)
X = np.linspace(-3, 3, 50).reshape(-1, 1)
y = X.squeeze()**2 + np.random.normal(0, 1, X.shape[0])  # Quadratic relationship with noise

# Train a Linear Regression model (even though the data is quadratic)
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Compute residuals
residuals = y - y_pred

# Plot residuals
plt.scatter(X, residuals, color="red", alpha=0.6)
plt.axhline(y=0, color='black', linestyle='dashed')  # Reference line at y=0
plt.xlabel("X values")
plt.ylabel("Residuals")
plt.title("Residual Plot - Poor Fit for Linear Regression")
plt.show()
```

### Fixing with issue with a Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Generate the same non-linear (quadratic) data
np.random.seed(42)
X = np.linspace(-3, 3, 50).reshape(-1, 1)
y = X.squeeze()**2 + np.random.normal(0, 1, X.shape[0])  # Quadratic relationship with noise

# Create a Polynomial Regression model (degree=2)
degree = 2  # Quadratic
poly_model = make_pipeline(PolynomialFeatures(degree), LinearRegression())

# Train the model
poly_model.fit(X, y)

# Make predictions
y_pred = poly_model.predict(X)

# Compute residuals
residuals = y - y_pred

# Plot residuals
plt.scatter(X, residuals, color="blue", alpha=0.6)
plt.axhline(y=0, color='black', linestyle='dashed')  # Reference line at y=0
plt.xlabel("X values")
plt.ylabel("Residuals")
plt.title("Residual Plot - Improved Fit with Polynomial Regression")
plt.show()
```
