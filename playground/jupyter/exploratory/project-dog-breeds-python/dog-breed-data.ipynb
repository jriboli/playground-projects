# Get familiar with the data

```python
# Imports
import pandas as pd

df_dogs = pd.read_csv("data\dog_breeds.csv")

df_dogs.head()
```

```python
# Can we find all the potential values for 'Character Traits' 

traits = df_dogs["Character Traits"]

list_of_traits = {}
for list in traits:
    individual_trait = str(list).split(",")
    for trait in individual_trait:
        trait = str(trait).strip().lower()
        if(list_of_traits.__contains__(trait)):
            list_of_traits[trait] = list_of_traits[trait] + 1
        else:
            list_of_traits[trait] = 1
        
list_of_traits

# Notes - Saw this - Need to understand how it works
# all_traits = set(trait for sublist in df["Traits"] for trait in sublist)
```

```python
# Alright now lets do something with these 'Traits' 

for trait in list_of_traits:
    print(trait)
    df_dogs[trait] = df_dogs["Character Traits"].apply(lambda x: trait in x)
    
df_dogs.head()
```

```python
# Lets try and graph out the Traits according to Countries

import matplotlib.pyplot as plot

#plot.hist(df_dogs, )
df_dogs["Country of Origin"].value_counts()
trait_categories = list_of_traits.keys()
trait_values = list_of_traits.values()

plot.bar(trait_categories, trait_values, color="skyblue")
plot.xticks(rotation=90)
plot.show()
```
